## 3. ***Data Wrangling***

Upon receiving the hotel booking dataset, we identified several notable issues that required our attention. The dataset contained 119,390 rows and 32 columns, but some critical columns like 'country,' 'agent,' and 'company' had incomplete data and duplicate entries, which hindered our ability to perform a thorough analysis. Therefore, we took immediate action to clean up the data and ensure the accuracy of our analysis.

To clean up the data, we need to remove any duplicate rows.

We should also handle any missing values that may be present.

 It's important to make sure that each column is in the appropriate datatype.
  
 Additionally, we should add two important columns: total guests and total nights stayed.

### Data Wrangling Code

```python
#Removing Duplicate valuse
Hotel_df[Hotel_df.duplicated()].shape
```

To guarantee precise data analysis, it's essential to identify and count the exact number of duplicated rows in the dataset.

```python
# Removing Duplicate values
Hotel_df.drop_duplicates(inplace=True)
Hotel_df.shape
```

Handling Missing values

```python
# Total number of missing values in the DataFrame
Hotel_df.isnull().sum()
```

```python
# I am correcting any missing or inaccurate data for the company or agent by accurately replacing it with the necessary value of 0.
Hotel_df['company'].fillna(0, inplace=True)
Hotel_df['agent'].fillna(0, inplace=True)

# Replacing the missing_values in country column with 'Unknown'
Hotel_df['country'].fillna('Unknown', inplace=True)
```

Null values were discovered in the 'company' and 'agent' columns for certain hotel bookings. In order to preserve data integrity, these missing values will be replaced with 0. Furthermore, the 'country' column has missing values that will be replaced with 'Unknown.

```python
Hotel_df['children'].fillna(Hotel_df['children'].mean(), inplace = True)
```

4 null values in children column will be replaced with the mean value of the column.

```python
#Checking the all null_values are removed or not
Hotel_df.isnull().sum()
```

It is clear, that the dataset being reviewed does not have any null values.

# checking the sum of adults, children and babies is zero or not

```python
Hotel_df[Hotel_df['adults']+Hotel_df['babies']+Hotel_df['children'] == 0].shape
```

We have identified 166 rows in which the total number of adults, children, and babies is zero. As a result, we will be removing these rows.

```python
Hotel_df.drop(Hotel_df[Hotel_df['adults']+Hotel_df['babies']+Hotel_df['children'] == 0].index, inplace = True)
```

 **Convert columns to appropriate datatypes**

```python
#Converting datatype of columns 'children', 'company' and 'agent' from float to int.
Hotel_df['children'] = Hotel_df['children'].astype(int)
Hotel_df['company'] = Hotel_df['company'].astype(int)
Hotel_df['agent'] = Hotel_df['agent'].astype(int)
```

```python
# Convert date columns to datetime data type
Hotel_df['reservation_status_date'] = pd.to_datetime(Hotel_df['reservation_status_date'])
```

```python
# Adding total_guests column = adults + children + babies
Hotel_df['total_guests'] = Hotel_df['adults'] + Hotel_df['children'] + Hotel_df['babies']
```

```python
# Adding total_nights_stayed = stayed in weekend nights + stayed in week nights
Hotel_df['total_nights_stayed'] = Hotel_df['stays_in_weekend_nights'] + Hotel_df['stays_in_week_nights']
Hotel_df.columns
```

### What all manipulations have you done and insights you found?

We analyzed hotel bookings using advanced data techniques to prepare the dataset for examination. Our approach involved several steps to ensure the accuracy and reliability of our data.

Firstly, we removed duplicate rows to focus on individual booking records and eliminate unnecessary redundancy. Secondly, we addressed the issue of missing data in columns such as 'country', 'agent', and 'company'. For 'country', we retained the missing values as they represented situations where customers did not specify a country during the booking process. However, for 'agent' and 'company', we filled in the gaps with 0 to show that those bookings were not made through any particular agent or company.

We converted the 'children', 'company', and 'agent' columns from float to integer data types since they represented whole numbers. This change allowed us to work with these columns more effectively in our later analyses.

We added two new columns, 'total guests' and 'total nights stayed', to acquire additional insights into booking patterns. By adding up the number of 'adults', 'children', and 'babies', we calculated the total number of guests for each booking. Similarly, we derived the overall nights stayed by combining 'stays_in_weekend_nights' and 'stays_in_week_nights'.

Our approach was highly effective, and we are confident in the accuracy and reliability of our dataset for further analysis.
